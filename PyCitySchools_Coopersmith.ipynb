```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = school_data_complete ["school_name"].nunique()
total_students = school_data_complete["student_name"].count()
total_budget = school_data ["budget"].sum()
average_math_score = school_data_complete ["math_score"]. mean()
average_reading_score = school_data_complete ["reading_score"].mean()
overall_passing_rate = (average_math_score + average_reading_score)/2
math_passing_score = ((school_data_complete ["math_score"] >=70)*100).mean()
reading_passing_score = ((school_data_complete["reading_score"] >=70)*100).mean()
district_summary_df = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "% Passing Math" : [math_passing_score],
    "% Passing Reading" : [reading_passing_score],
    "% Overall Passing Rate" : [overall_passing_rate]})
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].astype(float).map(
    "${:,.2f}".format)
district_summary_df["Total Students"] = district_summary_df["Total Students"].astype(float).map(
    "{:,.0f}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].astype(float).map(
    "{:,.2f}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].astype(float).map(
    "{:,.2f}".format)
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].astype(float).map(
    "{:,.2f}%".format)
district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].astype(float).map(
    "{:,.2f}%".format)
district_summary_df["% Overall Passing Rate"] = district_summary_df["% Overall Passing Rate"].astype(float).map(
    "{:,.2f}%".format)
district_summary_df                                                                                         
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
school_data_complete['spending_bins'] = pd.cut(school_data_complete['budget']/school_data_complete['size'], spending_bins, labels=group_names)
grouped_school = school_data_complete.groupby(['spending_bins'])
math = school_data_complete.groupby ('spending_bins')['math_score'].mean()
reading = school_data_complete.groupby ('spending_bins')['reading_score'].mean()
passing_math_spending = school_data_complete[school_data_complete['math_score']>=70].groupby(['spending_bins']).size()/grouped_school.size()*100
passing_reading_spending = school_data_complete[school_data_complete['reading_score']>=70].groupby(['spending_bins']).size()/grouped_school.size()*100
overall_passing_rate_spending =(passing_math_spending + passing_reading_spending)/2
scores_by_school_spending = pd.concat ([math,reading,passing_math_spending,passing_reading_spending,overall_passing_rate_spending], axis=1)
scores_by_school_spending_rename =scores_by_school_spending.rename(columns={"spending_bins": "Spending Ranges(Per Student),",
                                                "math_score": "Average Math Score",
                                                "reading_score": "Average Reading Score",
                                                0: "% Passing Math",
                                                1: "% Passing Reading",
                                                2: "% Overall Passing Rate"})
scores_by_school_spending_rename
scores_by_school_spending_rename["Average Math Score"] = scores_by_school_spending_rename["Average Math Score"].astype(float).map(
    "{:,.2f}".format)

scores_by_school_spending_rename["Average Reading Score"] = scores_by_school_spending_rename["Average Reading Score"].astype(float).map(
    "{:,.2f}".format)
scores_by_school_spending_rename["% Passing Math"] = scores_by_school_spending_rename["% Passing Math"].astype(float).map(
    "{:,.2f}%".format)
scores_by_school_spending_rename["% Passing Reading"] = scores_by_school_spending_rename["% Passing Reading"].astype(float).map(
    "{:,.2f}%".format)
scores_by_school_spending_rename["% Overall Passing Rate"] = scores_by_school_spending_rename["% Overall Passing Rate"].astype(float).map(
    "{:,.2f}%".format)
scores_by_school_spending_rename
```
